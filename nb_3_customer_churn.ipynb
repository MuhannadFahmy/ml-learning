```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset\Telco-Customer-Churn.csv")
```

```python
df = df.drop('customerID', axis='columns')
df.head(40)
```

```python
df.info()
```

```python
df = df[df.TotalCharges != " "]
```

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges)
```

```python
tenure_churn_no = df[df.Churn == 'No'].tenure
tenure_churn_yes = df[df.Churn == 'Yes'].tenure
plt.xlabel('Tenure')
plt.ylabel('Number of Customers')
plt.title('Customer Churn Visualiztion')
plt.hist([tenure_churn_yes, tenure_churn_no], color=['green', 'red'], label=['Churn=Yes', 'Churn=No'])
plt.legend()
```

```python
mc_churn_no = df[df.Churn == 'No'].MonthlyCharges
mc_churn_yes = df[df.Churn == 'Yes'].MonthlyCharges
plt.xlabel('Monthly Charges')
plt.ylabel('Number of Customers')
plt.title('Customer Monthly Charges Visualiztion')
plt.hist([mc_churn_yes, mc_churn_no], color=['green', 'red'], label=['Churn=Yes', 'Churn=No'])
plt.legend()
```

```python
df = df.replace('No internet service', 'No')
df = df.replace('No phone service', 'No')

for col in df:
    print(f'{col} : \t {df[col].unique()}')
```

```python
yes_no_col = {'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn' }

for col in yes_no_col:
    df[col] = df[col].replace({'Yes': 1, 'No': 0})
    
df['gender'] = df['gender'].replace({'Female': 1, 'Male': 0})
```

```python
df_dummies = pd.get_dummies(data=df, columns=['InternetService', 'Contract', 'PaymentMethod'], dtype=int)
```

```python
df_dummies.sample(4)
```

```python
df_dummies
```

```python
cols_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_dummies[cols_scale] = scaler.fit_transform(df_dummies[cols_scale])
```

```python
df_dummies.sample(5)
```

```python
X = df_dummies.drop('Churn', axis='columns')
Y = df_dummies.Churn
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
input = keras.Input((26,))

hidden1out = keras.layers.Dense(13, activation='relu')(input)
hidden2out = keras.layers.Dropout(0.2)(hidden1out)
hidden3out = keras.layers.Dense(5, activation='relu')(hidden2out)
# hidden4out = keras.layers.Dropout(0.2)(hidden3out)
output = keras.layers.Dense(1, activation='sigmoid')(hidden3out)

model = keras.Model(input, output)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
for i in range(len(y_pred)):
    y_pred[i] = 1 if y_pred[i] >= 0.5 else 0
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
import seaborn as sn

cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)

plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True, fmt ='d')
plt.xlabel('Prediction')
plt.ylabel('Truth')
```
